# Stage 1: Minimal Working SR (Swin2SR Inference Baseline)

This notebook tests our `model_swin2sr.py` architecture. It simulates loading a Low-Resolution (LR) Sentinel-2 patch, running it through the Swin2SR Vision Transformer, and verifying the 8x upscaled output.

```python
import sys
import os
sys.path.append(os.path.abspath("../src"))

import torch
import numpy as np
import matplotlib.pyplot as plt
from model_swin2sr import initialize_swin2sr_model, run_inference

print(f"PyTorch version: {torch.__version__}")
```

### 1. Initialize the Model
We initialize our Swin2SR wrapper. Since we don't have the specific Solafune `.pth` file downloaded yet, we will use random weights to prove the forward pass works.

```python
# Initialize model with 8x upscale
model = initialize_swin2sr_model()
print(f"Model upscaling factor: {model.upscale}x")
```

### 2. Generate a Mock Sentinel-2 Patch
Instead of a real image, we simulate a 64x64 pixel Low-Resolution (LR) tensor.

```python
# Create a dummy tensor representing a 64x64 RGB image
lr_tensor = torch.rand(1, 3, 64, 64)
print(f"Input LR Shape: {lr_tensor.shape}")
```

### 3. Run Inference
We pass the tensor through the model inside a `torch.no_grad()` block to save memory.

```python
# Run the super-resolution inference
sr_tensor = run_inference(model, lr_tensor)
print(f"Output SR Shape: {sr_tensor.shape}")

# Verify the math
assert sr_tensor.shape[-1] == lr_tensor.shape[-1] * 8
print("✅ Success! The model successfully upscaled the 64x64 image to 512x512 without crashing.")
```
